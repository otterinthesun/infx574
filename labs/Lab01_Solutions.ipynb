```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1. flipped 1

```python
def rot(alpha):
    a = alpha*np.pi/180
    return np.matrix([[np.cos(a), -np.sin(a)], [np.sin(a), np.cos(a)]])

def plotFigure(A):
    plt.figure(figsize=(8,8))
    plt.grid(b=True)
    min_limit = np.min(A) - np.max(A)*2.0
    max_limit = np.max(A) + np.max(A)*2.0
    plt.axes().set_aspect('equal')
    plt.axes().set_xlim((min_limit, max_limit))
    plt.axes().set_ylim((min_limit, max_limit))
    plt.plot(A[:,0], A[:,1])
    plt.axhline(0, color='black', linestyle = 'dashed')
    plt.axvline(0, color='black',linestyle = 'dashed')

A = np.matrix('0 0; 0 1; 0.5 0.5')
```

```python
plotFigure(A @ rot(30))
```

```python
plotFigure(A @ rot(-225))
```

```python
plotFigure(A @ rot(-12))
```

## 2 a new object

```python
p = (-90 + np.arange(11)*360/10)*np.pi/180
r = [2,1]*5
r = np.array(r + [2])
r[2] = 3
r[3] = 0.5
P = np.column_stack((r*np.cos(p), r*np.sin(p)))

def plotFigure1(A):
    plt.figure(figsize=(8,8))
    plt.axes().set_aspect('equal')
    plt.axes().set_xlim((-3, 3))
    plt.axes().set_ylim((-3, 3))
    plt.plot(A[:,0], A[:,1])
```

## 3 rotations

```python
plotFigure(P)
```

```python
plotFigure(P @ rot(-30))
```

## 4 Crazy Hat

```python
ch = pd.read_table('crazy_hat.tsv')
ch
```

```python
## 5 crazy hat plot

def plotCH(A):
    plt.figure(figsize=(8,8))
    plt.grid(b=True)
    min_limit = np.min(A) - np.max(A)*2.0
    max_limit = np.max(A) + np.max(A)*2.0
    plt.axes().set_aspect('equal')
    plt.axes().set_xlim((min_limit, max_limit))
    plt.axes().set_ylim((min_limit, max_limit))
    for g in ch.group.unique():
        i = ch.group == g
        plt.plot(A[i,0], A[i, 1], c='b')

    plt.axhline(0, color='black', linestyle = 'dashed')
    plt.axvline(0, color='black',linestyle = 'dashed')
    

chm = ch[['x', 'y']].as_matrix()
        
plotCH(chm)
```

```python
print(chm @ rot(-145))

plotCH(chm @ rot(-145))
```

## 6 flipx

```python
flipx = np.matrix('-1 0; 0 1')
plotCH(chm @ flipx)
```

## 7 stretchy

```python
def stretchy(s):
    return np.matrix([[1, 0], [0, s]])
plotCH(chm @ stretchy(0.3))
```

## 8 combined

```python
plotCH(chm @ rot(60) @ flipx @ rot(-70) @ stretchy(0.4) @ rot(30))
```
