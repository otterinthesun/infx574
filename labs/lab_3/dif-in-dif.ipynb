```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import random
%matplotlib inline
```

## Cross-sectional Estimator

```python
n = 1000
T = np.random.choice(2, n)
eps = 0.5*np.random.normal(size=n)
y = 0.5 + 2*T + eps
print(T[:20])
y[:30]
```

```python
df = pd.DataFrame({"y" : y, "T" : T })
print(df.head(10))
```

```python
plt.figure(figsize=(15,10))
plt.scatter(df['T'] + 0.2*np.random.uniform(size=n), df['y'])
```

```python
df.groupby("T").mean()
```

```python
m = smf.ols(formula = "y~T", data=df).fit()
print(m.summary())
```

## Before-after

```python
n = 1000
time = np.random.choice(2, n)
eps = 0.5*np.random.normal(size=n)
jobs = 0.5 + 2*(time > 0) + eps
jobs[:30]
```

```python
tc = pd.DataFrame({"jobs" : jobs, "time" : time })
print(tc.head(10))
```

```python
plt.figure(figsize=(15,10))
plt.scatter(tc['time'] + 0.2*np.random.uniform(size=n), tc['jobs'])
```

```python
m = smf.ols(formula = "jobs~time", data=tc).fit()
print(m.summary())
```

## Dif-in-dif estimator

```python
n = 1000 # set the number of observations
country = np.random.choice(['us','ca'], n) 
# pick random choices in n and assign values to the random selected result
print(country[:10]) # print the first 10 random selected values
#time = np.random.choice(['b', 'a'], n)
time = np.random.choice(2, n) #random select the time before as 0 and after as 1

eps = 0.5*np.random.normal(size=n)

jobs = 0.5 + 2*(country == 'us') + 0.8*(time == 1) + 0.5*(time == 1)*(country == 'ca') + eps
didDf = pd.DataFrame({"jobs" : jobs, "country" : country, 'time':time })
didDf.head()
```

```python
didDf.groupby(['country', 'time']).mean() 
# create the didff table with groupby function
```

the before 0 and after 1 for CA is 0.49 and 1.79, diff is 1.3 <br>
the before 0 and after 1 for US is 2.6 and 3.3, diff is 0.7 <br>
diff_in_diff = 1.3 - 0.7 = 0.6

```python
m = smf.ols(formula = "jobs ~ country + time + time*country", data=didDf).fit()
print(m.summary())
```
