```python
# define the gradient_AS_1D function
# x initial x
# R learning rate
# eps prameter ε

def Gradient_AS_1D(x, R, eps, bail_out_iteration):
    def f(x): #define fx
        return -x**2
    def grad(x): # define gradient x
        return -2*x
    
    print("x: ", x) #print initial x
    print("f(x): ", f(x)) #print initial fx
    print ("∇f(x): ", grad(x), "\n") #print initial gradx
    
    count = 0 # set initial iteration to 0
    while abs(grad(x)) > eps:
        x_new = x + R * grad(x) # calculating the new x using the x+ R * grad(x) function
        
        print("x_new: ", x_new) #print new x
        print("f(x_new): ", f(x_new)) #print new fx
        print ("∇f(x_new): ", grad(x_new)) #print new gradx
        print ("iteration: ", count+1, "\n") #print iteration number
        count +=1
        x = x_new

Gradient_AS_1D(1, 0.1, 1e-7, 500)
```

```python
#define fx and gradient
def f(x,A): #define f(x,A)
    return -x.T@A@x
def grad(x,A): # define gradient(x,A)
    return -2*A@x

print("x0:", "\n", x0, "\n")
print("A:", "\n", A, "\n")
print("f(x0,A): ", "\n", f(x0,A), "\n")
print("∇f(x0,A): ", "\n", grad(x0,A), "\n")


# define the gradient_AS_2D function
def Gradient_AS_2D(x, A, R, eps, bail_out_iteration): 
# x initial x
# A initial A
# R learning rate
# eps prameter ε   
    count = 0
    fx = f(x,A)
    gradient = grad(x,A)
    xlist = np.array([]).tolist()
    xlist.append(x)
    # get initial iteration
    print("x: ", x) 
    print("A: ", A)
    print("f(x, A): ", fx) 
    print ("∇f(x, A): ", gradient, "\n") 
    
    # for the second and beyond iteration
    while count < bail_out_iteration:
        if (abs(np.linalg.norm(gradient))) < eps:
            break
        x_new = x + R * gradient # calculating the new x using the x+ R * grad(x) function
        gradient = grad(x_new, A)
        xlist.append(x_new)
        fx = f(x_new,A)
        count +=1
        x = x_new

        print("x_new: ", x_new) #print new x
        print("f(x_new, A): ", fx) #print new fx
        print ("∇f(x_new, A): ", gradient) #print new gradx
        print ("iteration: ", count+1, "\n") #print iteration number
    return xlist, count

xlist = Gradient_AS_2D(x0, A, 0.1, 1e-6, 500)[0]
count = Gradient_AS_2D(x0, A, 0.1, 1e-6, 500)[1]

Gradient_AS_2D(x0, A, 0.05, 1e-7, 500)
```

#### 2. For R times create sample of N of standard normals, and take the mean of it. You'll have R means of N standard normals.
#### 3. Compute the mean-of-means, it's standard deviation, and 95% con dence region.
The 95% con dence region is the opposite of computing the percentage of observations that  t into the interval. The easiest way to  nd it is using sample quantiles. Check out the quantile function in R, or np.percentile function in python. For instance, np.percentile(x, 10) will give the 10th percentile q10 of an array x. 10th percentile is such a value that 10% of the sample values are smaller than that value.

```python
def normal_distri(R, N):
    mus=[] # create a list of mean
    for i in range(R):
        n = np.random.normal(size = N) # find random normal from N
        mus.append(np.mean(n)) # add the new random normal to the means list
    mu = np.mean(mus) # find the mean of the means list
    std = np.std(mus) # get the std for the means
    ci_lower = mu - 1.96*std # get the lower range of the 95% confidence interval
    ci_upper = mu + 1.96*std # get the upper range of the 95% confidence interval
    ci_low = np.percentile(mus, 2.5)
    ci_up = np.percentile(mus, 97.5)
    
    print("R: ", R, "\n")
    print("N: ", N, "\n")
    print("mu of the means list: ", mu, "\n")
    print("standard deviation of the means list: ", std, "\n")
    print("95% ci: ", "[", ci_lower, ", ", ci_upper, "]", "\n")
    print("95% ci percentile: ", "[", ci_low, ", ", ci_up, "]", "\n")
    
normal_distri(R,N)
```

### 2.2 Pareto distribution
Normal distribution has very nice properties, despite it's density function looking a little bit . . . nasty. But there are many things in our world that are not well approximated by normal distribution. Ex- amples include size of cities, links to webpages, popularity of actors, number of citations of researchers, size of wild res. . . . All of these are very inequal there are cities that are enourmous, but most of towns are small. Some researchers have hundreds of thousands of citations but most of them only a few. This kind of outcomes can be described by Pareto distribution. There are many ways to de ne it, but let's do it like this:
<br> __F(x)=1− x and f(x)=kxk0x−(k+1) wherex >= x0 andk>0.__
<br> The parameter x0 (called location) tells where the distribution  begins  and k describes how fast it falls for large x values (called shape). The expected value for Pareto random variable is
<br> __EX= k/(k-1)* x0 for k>1.__
<br> Note that for k   1 the expected value does not exist! Your task is to explore what the property means in practice.
<br> Set x0 = 1 and pick a k   1. Note: depending on the software you use, it may parameterized di erently and you may have to adjust the parameter choice accordingly. If you use np.random.pareto (python) or VGAM::rpareto (R), the parameterization is as described here.
