```python
def f(x):
    return - x**2
def grad(x): # define gradient of f(x)
    return -2 * x
```

```python
def gradient_ascent_1D(x0):
    print("x0:",x0)
    print("f(x0):",f(x0))
    print("∇f(x0):", grad(x0),"\n")

    R = 0.1 # learning speed
    eps = 10**(-6)
    max_iter = 100

    i = 0
    while True:
        if i > max_iter:
            break
        elif abs(grad(x0) - eps) < 0 :
            break
        else:
            x1 = x0 + R * grad(x0)
            x0 = x1
            i += 1
            print("x1:",x1)
            print("f(x1):",f(x1))
            print("∇f(x1):", grad(x1))
            print("iteration:",i,"\n")
            

gradient_ascent_1D(1) #pick an initial value at x0 = -1, which is smaller than the max x value
```
